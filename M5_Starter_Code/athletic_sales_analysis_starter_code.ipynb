### 1. Combine and Clean the Data
#### Import CSVs

```python
# (Read the CSV files into DataFrames.)   DOING THIS SECOND BC I HAVE TO IMPORT DATA FIRST!!!!!!!

#I AM IMPORTING THE DATA!!!!!!!!!! :D
athletic_sales_2020_path = Path('Resources/athletic_sales_2020.csv')
athletic_sales_2021_path = Path('Resources/athletic_sales_2021.csv')


# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv(athletic_sales_2020_path)
athletic_sales_2021_df = pd.read_csv(athletic_sales_2021_path)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
#combined_sales = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis=0).reset_index(drop=True)
combined_sales = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df])
combined_sales.reset_index(drop=True, inplace=True)


combined_sales
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

# Group by region, state, and city and calculate the sum of products sold
region_sold_most_prod = combined_sales.groupby(['region', 'state', 'city']).agg({'units_sold': 'sum', 'total_sales': 'sum'})

# Rename the sum column to "Total_Products_Sold"
region_sold_most_prod = region_sold_most_prod.rename(columns={'units_sold': 'Total_Products_Sold'})

# Display the resulting DataFrame
print(region_sold_most_prod)


# Show the top 5 results.
region_sold_most_prod = region_sold_most_prod.sort_values("Total_Products_Sold", ascending = False)
region_sold_most_prod.head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
region_sold_most_prod = region_sold_most_prod.pivot_table(values='Total_Products_Sold', index=['region', 'state', 'city'], aggfunc='sum')


# Rename the "units_sold" column to "Total_Products_Sold"
region_sold_most_prod = region_sold_most_prod.rename(columns={'units_sold': 'Total_Products_Sold'})


# Show the top 5 results.
region_sold_most_prod = region_sold_most_prod.sort_values("Total_Products_Sold", ascending = False)
region_sold_most_prod.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"

# Group by region, state, and city and calculate the sum of products sold
region_most_sales = combined_sales.groupby(['region', 'state', 'city']).agg({'total_sales': 'sum'})

# Rename the sum column to "Total Sales"
region_most_sales = region_most_sales.rename(columns={'total_sales': 'Total Sales'})

# Display the resulting DataFrame
print(region_most_sales)


# Show the top 5 results.
region_most_sales = region_most_sales.sort_values("Total Sales", ascending = False)
region_most_sales.head(5)

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
region_most_sales = combined_sales.pivot_table(values='total_sales', index=['region', 'state', 'city'], aggfunc='sum')


# Optional: Rename the "total_sales" column to "Total Sales"
region_most_sales = region_most_sales.rename(columns={'total_sales': 'Total Sales'})


# Show the top 5 results.
region_most_sales = region_most_sales.sort_values("Total Sales", ascending = False)
region_most_sales.head(5)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = women_athletic_sales_by_retailer.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales = daily_sales.sort_values("Total Sales", ascending = False)
daily_sales
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly_pivot = women_athletic_sales_by_retailer.resample('W').sum()


# Sort the resampled pivot table in ascending order on "Total Sales".
weekly_pivot = weekly_pivot.sort_values("Total Sales", ascending = False)
weekly_pivot
```
